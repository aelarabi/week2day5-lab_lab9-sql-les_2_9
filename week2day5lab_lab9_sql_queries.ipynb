2) Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week

```python
query='''SELECT rental_date,
CASE
WHEN DAYNAME(rental_date) = 'Monday' then 'week_day'
WHEN DAYNAME(rental_date) = 'Tuesday' then 'week_day'
WHEN DAYNAME(rental_date) = 'Wednesday' then 'week_day'
WHEN DAYNAME(rental_date) = 'Thursday' then 'week_day'
WHEN DAYNAME(rental_date) = 'Friday' then 'week_day'
ELSE 'Week_end'
END AS 'day_type'
FROM rental;'''
rentalday=pd.read_sql_query(query,engine)
display(rentalday)
```

5) How many films include Behind the Scenes content?

```python
query='''SELECT count(special_features) FROM film
WHERE special_features like '%%Behind the Scenes%%';'''
BTS=pd.read_sql_query(query,engine)
display(BTS)
```

6) Which kind of movies (rating) have a mean duration of more than two hours?

```python
query='''select rating, round (avg(length),2) as 'Rating with average film run time of more than 2 hrs' from film
group by rating 
having avg(length)>120
order by avg(length) asc;'''
runtime=pd.read_sql_query(query,engine)
display(runtime)
```

7) Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query='''select title, length,Rank() Over (order by length) as runtime_rank from film
WHERE (length IS NOT NULL) AND (length <> '  ')
group by length;'''
ranking=pd.read_sql_query(query,engine)
display(ranking)
```
